### Assumptions
1. No outliers in test dataset --> train set was representative enough for test set
2. Daily differences are negligible: I've compressed everything into 14 days of features

```python
# Data manipulation
import pandas as pd 
import numpy as np

# Loading Model
import pickle

# For Prediction
from sklearn.metrics import mean_squared_error, r2_score
```

### Reading Data

```python
# Load your data here (assume already split into test X and test Y:
# Assume test_X dataset has exactly 14 days worth of data points
df_x = pd.read_csv('data/yourXdata.csv')
df_y = pd.read_csv('data/yourYdata.csv')

oneHot_time_x = pd.read_csv('data/oneHotXdata.csv', index_col = 'Unnamed: 0')
oneHot_time_y = pd.read_csv('data/oneHotYdata.csv', index_col = 'Unnamed: 0')
df_x.head()
```

### Pre-processing

```python
# Convert timestamp to float

df_x['timestamp_float'] = pd.to_datetime(df_x['timestamp'], format='%H:%M').dt.hour + pd.to_datetime(df_x['timestamp'], format='%H:%M').dt.minute/60.0
df_y['timestamp_float'] = pd.to_datetime(df_y['timestamp'], format='%H:%M').dt.hour + pd.to_datetime(df_y['timestamp'], format='%H:%M').dt.minute/60.0

df_x = df_x.drop(['timestamp'], axis=1)
df_y = df_y.drop(['timestamp'], axis=1)
```

```python
# Assume all possible geo-hashes are there in X
# This cell converts our data to a m by n matrix, 
# where m is the total number of possible geohashes 
# and n is the number of timestamps sequentially ordered
# Extracting information by time-stamp to get features by timestamp
# Create list, then keep appending to it to form a list of lists

oneHot_x_time_col = oneHot_time_x.columns.to_list() # artificial df columns to list
oneHot_geohash_col = oneHot_time_x.T.columns.to_list()
testX_masterList = []

timeIndex = 0.0
dayIndex = 1
#dataIndex = 0

while (dayIndex <= 14):
    timeIndex = 0
    while(timeIndex < 24): # to parse every timestamp possible   
        subListY = [0]*len(oneHot_time_x.T.columns)
        df_short = df_x[(df_x.day == dayIndex) & (df_x.timestamp_float == timeIndex)]
        for row in df_short.itertuples():        
            indexCol = oneHot_geohash_col.index(row.geohash6)
            elemY = row.demand
            
            subListY[indexCol] = elemY

            #oneHot_transposed.at[indexCol,indexRow] = row.demand
        print("Time Done", timeIndex)
        timeIndex += 0.25
        testX_masterList.append(subListY)
    print("Day Done", dayIndex)
    dayIndex += 1
```

```python
# Extracting information by time-stamp to get features by timestamp
# Create list, then keep appending to it to form a list of lists

oneHot_y_time_col = oneHot_time_y.columns.to_list() # artificial df columns to list

testY_masterList = []

timeIndex = 0.0
timeIndex = 0
#dataIndex = 0

while(timeIndex < 1.25): # to parse every timestamp possible   
    subListY = [0]*len(oneHot_time_x.T.columns)
    df_short = df_y[df_y.timestamp_float == timeIndex]
    for row in df_short.itertuples():        
        indexCol = oneHot_geohash_col.index(row.geohash6)
        elemY = row.demand

        subListY[indexCol] = elemY

        #oneHot_transposed.at[indexCol,indexRow] = row.demand
    print("Time Done", timeIndex)
    timeIndex += 0.25
    testY_masterList.append(subListY)
```

#### Putting it into a DF to verify our matrix manipulation

```python
df_x_final = pd.DataFrame(testX_masterList, columns=oneHot_geohash_col)
df_x_final = df_x_final.T
df_x_final.columns = oneHot_x_time_col

df_x_final.tail()
df_x_final.shape
```

```python
df_y_final = pd.DataFrame(testY_masterList, columns=oneHot_geohash_col)
df_y_final = df_y_final.T
df_y_final.columns = oneHot_y_time_col

df_y_final.tail()
df_y_final.shape
```

Hopefully, everything checks out and we get True values for the previous 2 cells. Let us then proceed to prediction

### Prediction

```python
def reshape_data(x_test, y_test):

    # Need to transpose this list to the form 1329 * 1344
    x_test = x_test.transpose()
    x_test = x_test.reshape(1329,1344)

    #drop 1st 4 columns to make it divisible by 5
    x_test = x_test[:,4:]

    #reshape it to form same shape as y
    x_test = x_test.reshape(5*1329, -1)

    # Need to transpose this list to the form 1329 * 5
    y_test = y_test.transpose()

    # Convert it to a nx1 vector for LGBM (doesn't support multi-output)
    y_test = y_test.ravel()
    
    return x_test, y_test
```

```python
def report_results(model, x_test, y_test):
    y_pred = model.predict(x_test)
    print(y_pred)
    print(y_test)
    # cut-off
    print(np.shape(x_test))
   # Test RMSE error
    test_rmse = np.sqrt(mean_squared_error(y_test, y_pred))

   # Train fit
    test_r2 = r2_score(y_test, y_pred)


    print("Test RMSE: {:0.4f}".format(test_rmse))
    print("Test R2 score: {:0.4f}".format(test_r2))
    print("========================================================================")
    
    return test_rmse, test_r2, y_pred, y_test
```

```python
# Load my model
loaded_model = pickle.load(open("bestModel.pickle.dat", "rb"))

x_test, y_test = reshape_data(np.array(testX_masterList), np.array(testY_masterList))

# Next, to compute mean statistics
test_rmse, test_r2, y_pred, y_test = report_results(loaded_model, x_test, y_test)
```
